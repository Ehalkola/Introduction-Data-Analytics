a) Create pairplot of the data


```python
# Create variable to store information from dataset "penguins"
penguins = sns.load_dataset("penguins")

# Create a pair plot
sns.pairplot(penguins)
```

What correlations can you immediately see?

From this pair plot we can immidiately see: initial insights into distribution of the data, correlations between variables and  potential outliars.


Bonus extra task: check out the correlation matrix for this
dataset too 

```python
# Select numeric columns for computing correlation
numeric_columns = penguins.select_dtypes(include=["number"])

# Compute the correlation matrix
corr_matrix = numeric_columns.corr()

# Plot the correlation matrix as a heatmap using Seaborn
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
```

```python
# Create Scatter plot for bill and flipper length_mm hue island
sns.boxplot(x="species", y="flipper_length_mm", hue="island", data=penguins)
plt.title("Box Plot: Flipper Length by Species and Island")
plt.xlabel("Species")
plt.ylabel("Flipper Length (mm)")
plt.legend(title="Island")
plt.show()
```

```python
# Create box plot for bill length by island
sns.boxplot(x="species", y="bill_length_mm", hue="island", data=penguins)
plt.title("Box Plot: Bill Length by Species and Island")
plt.xlabel("Species")
plt.ylabel("Bill Length (mm)")
plt.legend(title="Island")
plt.show()
```

```python
# Box plot for bill_length_mm
sns.boxplot(x="species", y="body_mass_g", hue="island", data=penguins,)
plt.title("Box Plot: Body Mass by Species and Island")
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")
plt.legend(title="Island")
plt.show()
```

Box plot is best suited for providing insights into the spread of data, offering a clear understanding of central tendencies and variability across different factors, especially between categories such as species. It provides valuable insights into the ranges between different variables and is particularly suitable for comparing two variables and visualizing the ranges between data points effectively.

Part f) Violin plot for data, here you can see swarm plot created for flipper_length, bill length and bill body mass in this order

```python
# Violin plot for flipper_length_mm
plt.figure(figsize=(10, 8))
sns.violinplot(x='species', y='flipper_length_mm', hue='island', data=penguins, split=True)
plt.title('Violin Plot: Flipper Length vs Species (Hue: Island)')
plt.xlabel('Species')
plt.ylabel('Flipper Length (mm)')
plt.legend(title='Island')
plt.show()
```

```python
# Violin plot for bill_length_mm
plt.figure(figsize=(10, 8))
sns.violinplot(x='island', y='bill_length_mm', hue='species', data=penguins, split=True)

```

```python
# Violin plot for body_mass_g
plt.figure(figsize=(10, 8))
sns.violinplot(x='species', y='body_mass_g', hue='island', data=penguins, split=True)
plt.title('Violin Plot: Body Mass vs Species (Hue: Island)')
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')
plt.legend(title='Island')
plt.show()
```

Violin plot provides the best insight into data distribution and density, offering a detailed view compared to box or swarm plots. It is particularly effective for visualizing multiple variables and showing variations across these variables in a clear and comprehensive manner. Additionally, it is well-suited for highlighting differences between categories such as species in a dataset.

Part f) Swarm plot for data, here you can see swarm plot created for flipper_length, bill length and bill body mass in this order

```python
# Swarm plot for flipper_length_mm
plt.figure(figsize=(10, 8))
sns.swarmplot(x='species', y='flipper_length_mm', hue='island', data=penguins)
plt.title('Swarm Plot: Flipper Length vs Species (Hue: Island)')
plt.xlabel('Species')
plt.ylabel('Flipper Length (mm)')
plt.legend(title='Island')
plt.show()
```

```python
# Swarm plot for bill_length_mm
plt.figure(figsize=(10, 8))
sns.swarmplot(x='species', y='bill_length_mm', hue='island', data=penguins)
plt.title('Swarm Plot: Bill Length vs Species (Hue: Island)')
plt.xlabel('Species')
plt.ylabel('Bill Length (mm)')
plt.legend(title='Island')
plt.show()
```

```python
# Swarm plot for body_mass_g
plt.figure(figsize=(10, 8))
sns.swarmplot(x='species', y='body_mass_g', hue='island', data=penguins)
plt.title('Swarm Plot: Body Mass vs Species (Hue: Island)')
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')
plt.legend(title='Island')
plt.show()
```

Swarm plot is best suited for displaying individual data points within categories. It offers a clear visualization of how data points are distributed across different categories, providing valuable insights into the impactful variables within those categories.
