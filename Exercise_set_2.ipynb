<h2>Exercise 3: Data from salaries in India CSV</h2>

```python
# Read the csv-file(data_salaries_india.csv)
df = pd.read_csv('data_salaries_india.csv')
```

```python
# Get the columns names
df.columns
```

```python
#Convert all salaries into yearly salary
def yearly_wage(row):
    period = row['Salary'][-2:]
    number = int(''.join(filter(str.isdigit, row['Salary'])))
    if period == 'hr':
        number = int(number * 2117.01)
    elif period == 'mo':
        # months to year
        number = int(number * 12)
    return number
```

```python
# Remove the Indian rupees sign
df['Salary'] = df['Salary'].str.replace('₹', '' '')
df['Salary']
```

```python
# Remove characters that aren't numbers & convert to float
df['Salary'] = df['Salary'].str.replace('[^\d.]', '', regex=True).astype(float)
```

```python
# Print the head the dataframe
df.head()
```

```python
# Check for any missing values
missing_values = df.isna().sum()
print('Number of missing values in each column:', missing_values)
```

```python
# Remove rows with missing 'Company Name' values
df = df.dropna(subset=['Company Name'])
df
```

```python
# Find the most common values for job titles, company names, locations and salaries
job_title_counts = df['Job Title'].value_counts().head(5)
print(job_title_counts)

company_counts = df['Company Name'].value_counts().head(5)
print(company_counts)

location_counts = df['Location'].value_counts().head(5)
print(location_counts)

salary_counts = df["Salary"].value_counts().head(5)
print(salary_counts)
```

```python
# Combine similar job titles
job_title_map = {
    'Machine Learning Data Associate': 'Machine Learning Associate',
    'Machine Learning Associate': 'Machine Learning Associate',
}
df['Job Title'] = df['Job Title'].map(job_title_map).fillna(df['Job Title'])
```

```python
# Analyze distribution to check the balance of the data
job_title_bal = len(job_title_counts) / len(df['Job Title'].unique())
company_bal = len(company_counts) / len(df['Company Name'].unique())
location_bal = len(location_counts) / len(df['Location'].unique())

# Display the information to the user
print(job_title_bal)
print(company_bal)
print(location_bal)
```

```python
# Calculate observations  for each category, and compare them to see if the data is balanced or imbalanced
job_title_counts = df['Job Title'].value_counts()
company_counts = df['Company Name'].value_counts()
location_counts = df['Location'].value_counts()
salary_counts = df['Salary'].value_counts()

if job_title_counts.std() < 0.1 and company_counts.std() < 0.1 and location_counts.std() < 0.1 and salary_counts.std() < 0.1:
    print('The data is balanced.')
else:
    print('The data is unbalanced.')
```

<h3>Here in this comparison, we can see that data is unbalanced due to the deviation between the data being huge.<h3>

```python
# Calculate Z score for the Salary column
df['Salary_zscore'] = (df['Salary'] - df['Salary'].mean()) / df['Salary'].std()
df['Salary_zscore']
```

```python
# Salary greater than z score of 3
df[df['Salary_zscore'] > 3]
```

```python
# Salary less than z score of -3.
df[df['Salary_zscore'] > -3]
```

```python
# Create new salary DataFrame based on Salary z score
new_salary_df = df[(df['Salary_zscore']< 3) & (df['Salary_zscore'] > -3)]
new_salary_df
```

<h4>Z score is one of standard measurements used to measure how far data point is differing from its average value, z-score of 3 means that data point is three deviations. These are outliers, which should be deleted. When setting z-score treshold to 3, mean will remove any data point that is three or more deviations away from average value(mean). This way we can eliminate extreme deviations in data. Let's see how many outliers we will remove from this data-analysis.<h4>

```python
# Factorize the Role column & add the new column into Data
label1, unique1 = pd.factorize(df['Role'], sort=False)
df['ManagerRole'] = label1
df
```

<h3>Total of 63 outliers were removed from the analysis of the dataset.<h3>

```python
# Calculate Spearman correlations
corr_managerrole_salary = df['ManagerRole'].corr(df['Salary'], method='spearman')
corr_managerrole_sr = df['ManagerRole'].corr(df['Salaries Reported'], method='spearman')

# Display calculated Spearman correlations
print("Correlation between Manager Role and Salary:", corr_managerrole_salary)
print("Correlation between Manager Role and Salaries Reported:", corr_managerrole_sr)
```

<h4> A positive correlation of 0.4301196078482581 between 'ManagerRole' and 'Salary' columns suggests a moderate correlation between the two columns. This means that while the numerical label representing 'ManagerRole' increases, there is a tendency for the salary to increase. However, the correlation is that strong (over 0.5), indicating that the relationship may not be linear or that other factors would influence salary.

On the other hand, a correlation of -0.2544565723460406 between 'ManagerRole' and 'Salaries Reported' columns suggests a weak negative correlation. This means that as the numerical label representing the 'ManagerRole' increases, there is a slight tendency for the reported salaries to decrease. Again, the correlation is not that strong (-0.5), this indicates that the relationship may not be linear or that other factors would influence on reported salaries.
